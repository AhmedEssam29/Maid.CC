```python
# Configure dark background style sheet
plt.style.use('dark_background')


# Get the value counts for the price_range column
price_range_counts = df['price_range'].value_counts()

# Get pie chart slice labels
pie_chart_slice_labels = price_range_counts.index.to_numpy()  # Convert index to NumPy array

# Get pie chart slice values
pie_chart_slice_values = price_range_counts.to_numpy()

# Define custom colors for slices
custom_colors = ['slateblue', 'slategrey', 'darkcyan', 'seagreen']

# Ensure enough colors are defined for all slices
if len(custom_colors) < len(pie_chart_slice_labels):
    print("Warning: Not enough colors provided for all slices. Using default colors.")
    custom_colors = plt.cm.get_cmap('tab20').colors[:len(pie_chart_slice_labels)]  # Use default colormap

# Create pie chart with customizations
wedgeprops = dict(width=0.7, edgecolor='w')  # Adjust wedge width and edge color
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels, autopct="%1.1f%%", colors=custom_colors[:len(pie_chart_slice_values)])

# Customize chart elements for aesthetics
plt.title("Price Range Distribution", color='white', fontsize=16)  # Set title properties
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart

# Display chart
plt.show()
```

```python
# Get the value counts for the price_range column
price_range_counts = df['price_range'].value_counts()

# Get bar chart labels and values
bar_chart_labels = price_range_counts.index.to_numpy()
bar_chart_values = price_range_counts.to_numpy()

# Define a color list for the bars (adjust colors as desired)
custom_colors = ['gold', 'royalblue', 'lightcoral', 'limegreen']

# Create a bar chart
plt.bar(bar_chart_labels, bar_chart_values, color=custom_colors[:len(bar_chart_labels)])

# Set chart title and labels
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Count")

plt.xticks([0, 1, 2, 3], bar_chart_labels) 
# Add class names (labels) on top of each bar with counts
for index, value in enumerate(bar_chart_values):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom', fontsize=12)  # Adjust offset and font size

# Rotate x-axis labels for better readability with many categories
plt.xticks(rotation=0, ha='right')

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Display chart
plt.show()
```

#### from this plots I can see that each class is well represented and there is no imbalance in the dataset

```python
import matplotlib.pyplot as plt

# Calculate the number of rows needed
num_rows = (len(df.columns) - 1) // 5 + 1

# Create subplots
fig, axs = plt.subplots(num_rows, 5, figsize=(20, 4*num_rows))

# Plot histograms for each column
for i, column in enumerate(df.columns):
    row_index = i // 5
    col_index = i % 5
    axs[row_index, col_index].hist(df[column], bins=20, color='skyblue', edgecolor='black')
    axs[row_index, col_index].set_title(f'Distribution of {column}')
    axs[row_index, col_index].set_xlabel(column)
    axs[row_index, col_index].set_ylabel('Frequency')

# Hide any empty subplots
for i in range(len(df.columns), num_rows * 5):
    row_index = i // 5
    col_index = i % 5
    axs[row_index, col_index].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
